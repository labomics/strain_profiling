```python
import dask.dataframe as dd
from dask import delayed
def read_and_label_csv(filename):
    df_csv = pd.read_csv(filename,dtype={'best_mean_nm':str})
    df_csv['filename'] = os.path.basename(filename).replace('.csv','').replace('stat.','')
    return df_csv
```

```python
fnames = glob.glob(f'{workdir}/stats/*.csv')
dfs = [delayed(read_and_label_csv)(fname) for fname in fnames]
df = dd.from_delayed(dfs)
```

```python
df.head()
```

```python
df = df.compute()
```

```python
df['genome'] = df.src.apply(lambda d: ''.join(d.split('-')[2].split('_')[:2]))
df['type'] = df['filename'].str[0]
```

```python
df.loc[:,'type'] = [x+'C' for x in df['type']]
```

```python
subdf = df[df['src'].str.contains('GCA_000166035.1')]
sns.boxplot(data=subdf,x='type',y='sum-cov')
a = subdf[subdf['type']=='HC']['sum-cov']
b = subdf[subdf['type']=='LC']['sum-cov']
s,p = mannwhitneyu(a,b)
print(p)
#sns.boxplot(data=subdf,x='type',y='mean_dep')
plt.title(f'GCA_000166035.1,p={p:.2e}',fontsize=15)
plt.ylabel('Genome Coverage',fontsize=15)
plt.xticks(fontsize=15)
plt.xlabel('')
```

```python
pltdf = df.groupby(['genome','type']).agg({'relative abundance':np.median}).reset_index()
pltdf.head()
h_dict = dict(pltdf[pltdf['type']=='H'].set_index('genome')['relative abundance'])
l_dict = dict(pltdf[pltdf['type']=='L'].set_index('genome')['relative abundance'])
newpltdf = pd.DataFrame()
newpltdf['H'] = pltdf['genome'].map(h_dict)
newpltdf['L'] = pltdf['genome'].map(l_dict)
newpltdf.head()
newpltdf = newpltdf.loc[(newpltdf['H']/newpltdf['L']).sort_values().index]
newpltdf['L'] = newpltdf['L']*-1
newpltdf.plot(kind='bar',stacked=True,figsize=(12,6))
plt.xticks([],[])
plt.xlabel('strains',fontsize=15)
plt.ylabel('relative abundance',fontsize=15)
```
